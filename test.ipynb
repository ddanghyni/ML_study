```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib

# 국가 이름 나이 성별 직업 데이터
data = {
    '국가': ['한국', '미국', '일본', '중국', '인도'],
    '나이': [25, 30, 35, 40, 45],
    '성별': ['남', '여', '남', '여', '남'],
    '직업': ['개발자', '디자이너', '마케터', '회계사', '영업']
}

df = pd.DataFrame(data)
df.head()


sns.set_theme(palette="pastel", style="darkgrid", font="AppleGothic") # 폰트 설정 in mac

plt.figure(figsize=(15, 7), dpi=300)
sns.barplot(x='국가', y='나이', data=df)

plt.show()
```
